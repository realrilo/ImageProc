```python
from PIL import Image, ImageFilter, ImageOps
import requests
from io import BytesIO
import matplotlib.pyplot as plt
import SimpleITK as sitk
import numpy as np
```

```python
# Open an image file
url = "https://res.cloudinary.com/dd4uyxyju/image/upload/v1727079232/Mask_Adult_Hero_MO_uppt3c.jpg"
response = requests.get(url)
image = Image.open(BytesIO(response.content))
```

```python
plt.imshow(image)
plt.axis('off')  # Hide axes
plt.show()
```

Blurred

```python
# Apply a filter to the image
filtered_image = image.filter(ImageFilter.BLUR)
```

```python
# Save the modified image
filtered_image.save('filtered_blurred.jpg')
```

```python
# Display the original and filtered images
image.show()
filtered_image.show()
```

```python
plt.imshow(filtered_image)
plt.axis('off')  # Hide axes
plt.show()
```

Rotate

```python
# Rotate the image by 45 degrees
rotated_image = image.rotate(180)

# Save the rotated image
#rotated_image.save('rotated_example.jpg')

# Display the rotated image
rotated_image.show()
```

```python
plt.imshow(rotated_image)
plt.axis('off')  # Hide axes
plt.show()
```

Sharpen

```python
# Apply sharpening filter
sharpened_image = image.filter(ImageFilter.SHARPEN)

# Save the sharpened image
#sharpened_image.save('sharpened_example.jpg')

# Display the sharpened image
sharpened_image.show()
```

```python
plt.imshow(sharpened_image)
plt.axis('off')  # Hide axes
plt.show()
```

greyscale

```python
# Convert the image to grayscale
grayscale_image = image.convert('1') # L

# Save the grayscale image
#grayscale_image.save('grayscale_example.jpg')

# Display the grayscale image
grayscale_image.show()
```

```python
plt.imshow(grayscale_image)
plt.axis('off')  # Hide axes
plt.show()
```

Colorize

```python
# Convert the image to grayscale
grayscale_image = image.convert('L')

# Apply colorize to the grayscale image
result_image = ImageOps.colorize(grayscale_image, 'red', 'white')

# Display the colorized image
result_image.show()
```

```python
plt.imshow(result_image)
plt.axis('off')  # Hide axes
plt.show()
```

Solarizing

```python
solarize = ImageOps.solarize(image)
plt.imshow(solarize)
plt.axis('off')  # Hide axes
plt.show()
```
